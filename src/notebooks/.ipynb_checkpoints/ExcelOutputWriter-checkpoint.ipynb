Initialize Excel File & write output over selected homes

```python


def addExcelTabs(workbook,df,dfElec,dfGas):

    #Loop through home DF for worksheets
    for i in range(len(df)):

        worksheet = workbook.add_worksheet("Site "+str(i+1))

        # Widen the first column to make the text clearer.
        worksheet.set_column('A:A', 20)

        # Add a bold format to use to highlight cells.
        bold = workbook.add_format({'bold': 1})
        topline = workbook.add_format()
        grey = workbook.add_format()
        grey.set_pattern(1)
        grey.set_bg_color('grey')

        # Write common text for header
        worksheet.write('A1', 'CalTest - Uncalibrated',bold)
        worksheet.write('A2', 'Site', bold)
        worksheet.write('B2', str(i+1), bold)

        #Vendor Table
        worksheet.write('A4', 'VENDOR FILL THIS TABLE',bold)
        worksheet.write('A5', 'Annual Energy Use', bold)
        worksheet.write('B4', 'Software Projection', bold)
        worksheet.write('B5', 'Before', bold)
        worksheet.write('C5', 'After', bold)
        worksheet.write('D5', 'Savings', bold)
        worksheet.write('E5', 'Realized', bold)
        worksheet.write('F5', 'Pass?', bold)
        worksheet.write('A6', 'Gas - total therms',bold)
        worksheet.write('A7', 'Electric - total kWh', bold)
        #worksheet.write('A8', 'Calibrated?', bold)

        worksheet.write('H4', 'Building Analysis Results', bold)
        worksheet.write('H5', 'Before', bold)
        #worksheet.write('H6', df['therms_before'])
        #worksheet.write('I6', df['therms_after'])
        #worksheet.write('H7', df['kwh_before'])
        #worksheet.write('I7', df['kwh_after'])
        #worksheet.write('J6', df['gas_savings'])
        #worksheet.write('J7', df['electricity_savings'])
        worksheet.write('I5', 'After', bold)
        worksheet.write('J5', 'Savings', bold)

        # Basic House Information.
        worksheet.write('A9', 'Basic House Information',bold)
        worksheet.write('A10', 'Year Built')
        worksheet.write('A11', 'Zip Code')
        worksheet.write('A12', 'CA Climate Zone')
        worksheet.write('A13', 'Conditioned Area')
        worksheet.write('A14', 'Volume')
        worksheet.write('A15', '# of Stories Above Grade')
        worksheet.write('A16', 'Ceiling Height')
        worksheet.write('A17', 'Bedrooms')
        worksheet.write('A18', 'Front Orientation')
        worksheet.write('A19', 'Foundation Type')
        worksheet.write('A20', 'Duct Location')

        worksheet.write('B10', df[vardict['year']][i])
        worksheet.write('B11', df['zipcode'][i])
        worksheet.write('B12', df['ca_climate_zone'][i])
        worksheet.write('B13', df['pre_conditioned_area'][i])
        worksheet.write('B14', df['volume'][i])
        worksheet.write('B15', df['stories_above_grade'][i])
        worksheet.write('B16', df['ceiling_height'][i])
        worksheet.write('B17', df['bedrooms'][i])
        worksheet.write('B18', df['front_orientation'][i])
        worksheet.write('B19', df['foundation_type'][i])
        worksheet.write('B20', df['duct_location'][i])

        #HVAC/Equiptment
        worksheet.write('A22', 'HVAC / Equipment',bold)
        worksheet.write('A23', 'Gas Furnace Efficiency')
        worksheet.write('A24', 'Central Air Conditioner Efficiency (SEER)')
        worksheet.write('A25', 'Cooling Capacity')
        worksheet.write('A26', 'Duct Insulation R (nominal)')
        worksheet.write('A27', 'Duct Leakage CFM25')
        worksheet.write('A28', 'Water Heater Type')

        worksheet.write('B22', 'Existing',bold)    
        worksheet.write('B23', df['gas_furnace_efficiency_existing'][i])
        worksheet.write('B24', df['central_ac_existing'][i])
        worksheet.write('B25', df['cooling_efficiency_existing'][i])
        worksheet.write('B26', df['duct_insulation_r_existing'][i])
        worksheet.write('B27', df['duct_leakage_existing'][i])
        worksheet.write('B28', df['water_heater_type_existing'][i])

        worksheet.write('C22', 'New',bold)
        worksheet.write('C23', df['gas_furnace_efficiency_new'][i])
        worksheet.write('C24', df['central_ac_new'][i])
        worksheet.write('C25', df['cooling_efficiency_new'][i])
        worksheet.write('C26', df['duct_insulation_r_new'][i])
        worksheet.write('C27', df['duct_leakage_new'][i])
        worksheet.write('C28', df['water_heater_type_new'][i])

        #Building Enclosure
        worksheet.write('A30', 'Building Enclosure',bold)
        worksheet.write('A31', 'Building Shell Leakage CFM50')
        worksheet.write('A32', 'Foundation Floor Insulation')
        worksheet.write('A33', 'Foundation Wall Insulation')
        worksheet.write('A34', 'Attic Type')
        worksheet.write('A35', 'Attic Ceiling Area')
        worksheet.write('A36', 'Attic Insulation')
        worksheet.write('A37', 'Wall Construction Type')
        worksheet.write('A38', 'Wall Insulation')
        worksheet.write('A39', 'Window Type/spec')

        worksheet.write('B30', 'Existing',bold)    
        worksheet.write('B31', df['shell_leakage_cfm50_existing'][i])
        worksheet.write('B32', df['foundation_floor_ins_existing'][i])
        worksheet.write('B33', df['foundation_wall_ins_existing'][i])
        worksheet.write('B34', df['attic_type_existing'][i])
        worksheet.write('B35', df['attic_ceiling_area_existing'][i])
        worksheet.write('B36', df['attic_insulation_existing'][i])
        worksheet.write('B37', df['wall_construction_existing'][i])
        worksheet.write('B38', df['wall_insulation_existing'][i])
        worksheet.write('B39', df['window_type_existing'][i])

        worksheet.write('C30', 'New',bold)
        worksheet.write('C31', df['shell_leakage_cfm50_new'][i])
        worksheet.write('C32', df['foundation_floor_ins_new'][i])
        worksheet.write('C33', df['foundation_wall_ins_new'][i])
        worksheet.write('C34', df['attic_type_new'][i])
        worksheet.write('C35', df['attic_ceiling_area_new'][i])
        worksheet.write('C36', df['attic_insulation_new'][i])
        worksheet.write('C37', df['wall_construction_new'][i])
        worksheet.write('C38', df['wall_insulation_new'][i])
        worksheet.write('C39', df['window_type_new'][i])

        #Other Retrofits
        worksheet.write('A41', 'Other Retrofits',bold)
        worksheet.write('A42', df['retrofit_measures_other_new'][i])

        #Dates & Weather
        worksheet.write('A44', 'Retrofit Start',bold)
        worksheet.write('A45', 'Retrofit Complete',bold)
        worksheet.write('A46', 'Weather Station Used',bold)

        worksheet.write('B44', df['retrofit_start_date'][i])
        worksheet.write('B45', df['retrofit_end_date'][i])
        worksheet.write('B46', df['weather_station'][i])

        #Add Energy Use Data
        #idvar = df['idvar']
        
        #Add raw elec use
        #elec = dfElec[dfElec['idvar'][i] == idvar]
        
        #for i 
        
        #Add raw gas use 
        #gas = dfGas
    

```
