```python
class WeatherSettings():
    """This class contains settings used in weather and PRISM
    """
    hdd_ref_temps = range(40,81)
    cdd_ref_temps = range(60,81)
    wu = False
    
    def __init__(self,hdd_ref_temps=range(40,81),cdd_ref_temps=range(60,81),wu=False,namedict={'usaf':'station_number', 'temps':'tempurature','date':'date','CliZn':'climate_zone','Lat':'lat','Lon':'long','Elevation':'elevation'}):
        
        self.hdd_ref_temps = hdd_ref_temps
        self.cdd_ref_temps = cdd_ref_temps
        self.namedict = namedict
        self.wu = wu
        pass
    
    
class WeatherData():
    """This class contains the logic for pulling and preparing weather data for PRISM given
        a weather dataset, and degree day settings object
    
    """
    data_set = ""
    impute = None
    hdd = None
    cdd = None
    hdd_ref_temps = range(55,65)
    cdd_ref_temps = range(65,75)
    wu = False
    wu_key="6bc1ff4991cf0aeb"
    
    def __init__(self,weather_data_set_name,weather_settings):
        
        self.data_set = weather_data_set_name
        self.hdd_ref_temps = weather_settings.hdd_ref_temps
        self.cdd_ref_temps = weather_settings.cdd_ref_temps
        
        if weather_settings.wu:
            self.data = self.get_weather_data(weather_settings)
        else:
            self.data = pd.read_csv(self.data_set)
        pass
        
        #clean and prepare data
        self.data = self.prepare(self.data, weather_settings.namedict)
        
        self.hdd = self.create_hdd_dataset(self.data,weather_settings)
        
        self.cdd = self.create_cdd_dataset(self.data,weather_settings)
        
        #self.normals = self.create_normals_dataset(self.hdd,self.cdd,weather_settings)
    
    @staticmethod
    def get_weather_data():
        return 0
    
    @staticmethod    
    def imput():
        return 0
    
    @staticmethod
    def prepare(weather_data, namedict={'usaf':'station_number', 'temps':'tempurature','date':'date','CliZn':'climate_zone','Lat':'lat','Lon':'long','Elevation':'elevation'}):
        """Takes a raw weather dataset and renames columns to standard column names
        
        standardized column names are:
        station_number, station_name, date, tempurature, climate_zone, long, lat, elevation
        """
        
        #rename columns to standard names
        weather_data = weather_data.rename(columns=namedict)
        
        #check date is pandas datetime
        weather_data['date']= pd.to_datetime(weather_data['date'])
        
        #set date as index
        weather_data = weather_data.set_index([weather_data['date']])
        
        return weather_data
        
    
    @staticmethod
    def create_hdd_dataset(weather_data,weather_settings):
        """Creates a new dataframe of cumulative heating degree days over period for a variety of reference tempuratures
        """
        
        #create a dataset of cumulative heated degree days
        #hdd = weather_data['station_number'].groupby('station_number')
        
        for hdd_ref_temp in weather_settings.hdd_ref_temps:
            weather_data['hdd'] = hdd_ref_temp - weather_data['tempurature']
            weather_data['hdd'] = weather_data['hdd'].apply(lambda x: 0 if x<0 else x)
            
            #create cumulative total
            #weather_data.index(['station_number','date'])
            #weather_data.sort(['station_number','date'])
            colname = 'hdd' + str(hdd_ref_temp)
            weather_data[colname] = weather_data.groupby(['station_number']).hdd.cumsum()
            
        return weather_data

    @staticmethod
    def create_cdd_dataset(weather_data,weather_settings):

        #create a dataset of cumulative heated degree days
        #hdd = weather_data['station_number'].groupby('station_number')
        
        for cdd_ref_temp in weather_settings.cdd_ref_temps:
            weather_data['cdd'] = cdd_ref_temp - weather_data['tempurature']
            weather_data['cdd'] = weather_data['cdd'].apply(lambda x: 0 if x<0 else x)
            
            #create cumulative total
            #weather_data.index(['station_number','date'])
            #weather_data.sort(['station_number','date'])
            colname = 'cdd' + str(cdd_ref_temp)
            weather_data[colname] = weather_data.groupby(['station_number']).cdd.cumsum()
            
        return weather_data
    
    @staticmethod
    def create_long_run_averages(weather_data,weather_settings):
        """Takes a dataframe of daily weather data by stations and returns longrun daily average
        
        """
        
        
        weather_data.date = pd.to_datetime(weather_data.date)

        #weather_lra_hdd = weather_data.groupby([weather_data['station_number'],weather_data.index.month,weather_data.index.day]).tempurature.mean()
        weather_lra_hdd = weather_data.groupby([weather_data['station_number'],weather_data.index.month,weather_data.index.day]).tempurature.mean()
        
        weather_lra_cdd
        
        #combine to make normals
        
        return weather_lra_hdd, weather_lra_cdd
    

    @staticmethod
    def save_to_database():
        return 0

    
```
